```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']

df1['Income Bank'] = pd.cut(df1['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df1['Income Bank'].value_counts().plot(kind = 'bar')
```

```python
#examine the distribution of unique categories in categorical columns
categorical_cols = df1[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee_Structure","Loyalty Classification",
                        "Properties Owned","Risk Weighting","Income Bank"]].columns

for col in categorical_cols:
    print("Value Counts for '{col':")
    display(df1[col].value_counts())
```

##Univariate Analysis

```python
for i, predictor in enumerate(df1):
    plt.figure(i)
    sns.countplot(data=df1, x=predictor, hue='GenderId')
```

##Bivariate Analysis

```python
predictors = ["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", 
              "Occupation", "Fee_Structure", "Loyalty Classification", 
              "Properties Owned", "Risk Weighting", "Income Bank"]
for i, predictor in enumerate(predictors):
    plt.figure(i)
    sns.countplot(data=df1, x=predictor, hue='Nationality')
```

```python
#Hisplot of value counts for different occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df1[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

#Numerical Analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings']
#univariate analysis and visualization
plt.figure(figsize=(8,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(2,1,i+1)
    sns.histplot(df1[col],kde=True)
    plt.title(col)
plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance']
#univariate analysis and visualization
plt.figure(figsize=(8,15))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,1,i+1)
    sns.histplot(df1[col],kde=True)
    plt.title(col)
plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts',
                  'Saving Accounts','Foreign Currency Account','Business Lending']
#univariate analysis and visualization
plt.figure(figsize=(20,20))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df1[col],kde=True)
    plt.title(col)
plt.show()
```

Heatmaps

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts',
                  'Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df1[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

##Insights of EDA
1. The strongest positive correlation occur among "Bank Deposits" with
